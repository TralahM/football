# Exploring the Data of the English premiere League 2017-2018 Season statistics as obtained from [football-data.co.uk](football-data.co.uk/englandm.php)

## To do this we are going to import some python libraries to facilitate this process

### That is:
- pandas
- matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

* Read the dataset for the 2017-2018 season located in ./data/E0_1718.csv using pandas.read_csv

```python
df1718=pd.read_csv("./data/E0_1718.csv")
```

* Let's check the shape of this dataset using df1718.shape

```python
df1718.shape
```

We see that the dataset has 380 rows or entries and 65 columns or features

* Now let us peek into the dataset and see the first five columns

```python
df1718.head(5)
```

#### How many goals are scored by the Home Team and Away Team on average?
Using the columns FTHG, and FTAG we calculate the average
With dataframe indexing we apply the mean function

```python
avg_fthg=df1718['FTHG'].mean()
avg_ftag=df1718['FTAG'].mean()
print("AVERAGE FULL TIME HOME GOALS: ",avg_fthg," AVERAGE FULL TIME HOME GOALS: ",avg_ftag)
```

* Clearly we see the home team scores more goals frequently than the away team.
* For every goal scored by the away team, how many goals are scored by the home team on average?

```python
ratio=avg_fthg/avg_ftag
print(ratio," : ",'1')
```

* Seems like the home team scores 33% more goals on average than the away team

* How many draws occurred on average in this season?

```python
ndraws=[i for i in range(380) if df1718['FTR'][i]=='D']
nhwins=[i for i in range(380) if df1718['FTR'][i]=='H']
nawins=[i for i in range(380) if df1718['FTR'][i]=='A']
print(len(ndraws)+len(nhwins)+len(nawins)==380) # Prove accuracy of the operation
print("Number of Draws: ",len(ndraws))
print("Number of Home Wins: ",len(nhwins))
print("Number of Away Wins: ",len(nawins))
```

* By the above statistics let's calculate the percentages thereof.

```python
phwins,pawins,pdraws=len(nhwins)/380,len(nawins)/380,len(ndraws)/380
print("Home wins: %f "%(phwins*100))
print("Draws: %f "%(pdraws*100))
print("Away wins: %f "%(pawins*100))
```

* So we see the Home Team won 46% of the games in this season, The away team 28% of them and 26% of the games ended in a draw

#### Let's Visualize this in a pie chart

```python
labels=['Home Team Wins','Draws','Away Team Wins']
sizes=[phwins*100,pdraws*100,pawins*100]
explode=(0,0.13,0)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title("A pie chart depicting the Full time results averages of the EPL 2017-2018 Season.")
```

* What about the half time results?

```python
# Calculating the number of events at Half time
nhtdraws=[i for i in range(380) if df1718['HTR'][i]=='D']
nhthwins=[i for i in range(380) if df1718['HTR'][i]=='H']
nhtawins=[i for i in range(380) if df1718['HTR'][i]=='A']
print(len(nhtdraws)+len(nhthwins)+len(nhtawins)==380) # Prove accuracy of the operation
print("Number of Draws: ",len(nhtdraws))
print("Number of Home Wins: ",len(nhthwins))
print("Number of Away Wins: ",len(nhtawins))

# Calculating Percentages
phthwins,phtawins,phtdraws=len(nhthwins)/380,len(nhtawins)/380,len(nhtdraws)/380
print("Home wins: %f "%(phthwins*100))
print("Draws: %f "%(phtdraws*100))
print("Away wins: %f "%(phtawins*100))

# Plotting the stats
labels=['Home Team Half Time Wins','Half Time Draws','Away Team Half Time Wins']
sizes=[phthwins*100,phtdraws*100,phtawins*100]
explode=(0,0.13,0)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title("A pie chart depicting the Half time results averages of the EPL 2017-2018 Season.")
```

#### Interesting it seems the number of draws at Half Time are more than the number of Home Team wins.
##### To see this more clearly let us plot the two charts side by side...

```python
fig2=plt.figure(figsize=(15,6))


axht=fig2.add_subplot(121)
labels=['Home HT Wins','HT Draws','Away Team HT Wins']
sizes=[phthwins*100,phtdraws*100,phtawins*100]
explode=(0,0.13,0)
axht.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
axht.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
axht.set_title('Half Time Stats.')

axft=fig2.add_subplot(122)
labels=['Home Team Wins','Draws','Away Team Wins']
sizes=[phwins*100,pdraws*100,pawins*100]
explode=(0,0.13,0)
axft.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
axft.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
axft.set_title('Full Time Stats.')

#plt.title("Comparing Half Time vs Full Time Results")
plt.subplots_adjust(wspace=0.5)
```

### Comparing Shots on Target with the number of goals scored in a game.
* Let us now compare the number of goals scored in the season with the number of shots on target.
* Is there a relationship between the number of shots on target with the number of goals scored in a game?

```python
goals_per_match=df1718['FTHG']+df1718['FTAG']
shots_target_per_match=df1718['HST']+df1718['AST']
len(goals_per_match)==len(shots_target_per_match)
```

```python
#### Let's Visualize these stats using a scatter plot and a line plot.
fig3=plt.figure(figsize=(13,5))
ax1=fig3.add_subplot(131)
ax1.scatter(shots_target_per_match,goals_per_match,color='r')
ax1.set_title("Scatter plot of goals scored vs shots on target ")
ax1.set_xlabel("Number of Shots on Target")
ax1.set_ylabel("Number of Goals Scored")
ax1.grid(True)

ax2=fig3.add_subplot(132)
ax2.plot(shots_target_per_match,goals_per_match)
ax2.set_title("Line plot of goals scored vs shots on target")
ax2.set_xlabel("Number of Shots on Target")
ax2.set_ylabel("Number of Goals Scored")
ax2.grid(True)

ax3=fig3.add_subplot(133)
ax3.bar(shots_target_per_match,goals_per_match,color='g')
ax3.set_title("Bar plot of goals scored vs shots on target")
ax3.set_xlabel("Number of Shots on Target")
ax3.set_ylabel("Number of Goals Scored")
ax3.grid(True)
plt.subplots_adjust(wspace=0.5)
```

* Clearly we see that there is no evident relationship between the number of shots on target and the number of goals scored in this season.
